## Pentagon Numbers



Pentagonal numbers are generated by the formula, $P_n=n(3n−1)/2.$ The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k − P_j|$ is minimised; what is the value of D?


```python
import pandas as pd
import scipy.optimize as opt
```

```python
def nthPentagonal(n):
    return n*(3*n - 1)*0.5

def nthPentagonalInv(x):
    if not isPentagonal(x):
        return False, None
    
    return True, ((24*x + 1)**0.5 + 1)/6

def isPentagonal(x):
    check = (24*x + 1)**0.5
    return check - round(check)
```

```python
def getSumAndDifference(n_j, n_k):
    if n_k < n_j:
        n_j, n_k = n_k, n_j

    p_j = nthPentagonal(n_j)
    p_k = nthPentagonal(n_k)
    # print(f"{p_k} - {p_j} = {d}")
    return p_k + p_j, p_k - p_j
```

```python
def objective(n):
    n_j, n_k = n[0], n[1]
    _, d = getSumAndDifference(n_j, n_k)
    return d

def constraint(n):
    n_j, n_k = n[0], n[1]
    s, d = getSumAndDifference(n_j, n_k)
    return isPentagonal(s), isPentagonal(d)

cons = [
    {"type": "eq", "fun": lambda n: constraint(n)[0]},
    {"type": "eq", "fun": lambda n: constraint(n)[1]},
    {"type": "ineq", "fun": lambda n: n[0] - n[1]}
]
```

```python
# TODO: Integer solutions only
# TODO
sol = opt.minimize(objective, (1, 4), method="SLSQP", constraints=cons, options={"disp": True})

print(sol.x)
print(sol.fun)
```
