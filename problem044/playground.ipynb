## Pentagon Numbers



Pentagonal numbers are generated by the formula, $P_n=n(3n−1)/2.$ The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k − P_j|$ is minimised; what is the value of D?


```python
def nthPentagonal(n, pentagonals_dict=None):
    try:
        return pentagonals_dict[n]
    except (KeyError, TypeError):
        return int(n*(3*n - 1)*0.5)

def isPentagonal(x, pentagonals_inv_dict=None):
    try:
        return pentagonals_inv_dict[x]
    except (KeyError, TypeError):
        check = ((24*x + 1)**0.5 + 1) / 6
        return check.is_integer()

def findSmallestDiff():
    pentagonals = dict()
    pentagonals_inv = dict()
    j = 2
    while True:
        p_j = nthPentagonal(j, pentagonals)
        if j not in pentagonals.keys():
            pentagonals[j] = p_j
            pentagonals_inv[p_j] = j
        for k in range(2, j):
            p_k = nthPentagonal(k, pentagonals)
            if k not in pentagonals.keys():
                pentagonals[k] = p_k
                pentagonals_inv[p_k] = k
            p_s = p_j + p_k
            # s = pentagonals_inv[p_s]
            if isPentagonal(p_s, pentagonals_inv):
                # d = pentagonals_inv[p_d]
                p_d = abs(p_j - p_k)
                if isPentagonal(p_d, pentagonals_inv):
                    # print(f"j: ({j}, {p_j}); k: ({k}, {p_k}); s: ({s}, {p_s}); d: ({d}, {p_d})")
                    print(f"j: ({j}, {p_j}); k: ({k}, {p_k}); s: ({p_s}); d: ({p_d})")
                    return p_d

        j += 1

def findSmallestDiffWithN(n=2400):
    pentagonals = dict()
    pentagonals_inv = dict()
    for x in range(n):
        p_x = nthPentagonal(x)
        pentagonals[x] = p_x
        pentagonals_inv[p_x] = x

    for j in range(2, n):
        p_j = pentagonals[j]
        for k in range(2, j):
            p_k = pentagonals[k]

            try:
                p_s = p_j + p_k
                p_d = abs(p_j - p_k)
                s = pentagonals_inv[p_s]
                d = pentagonals_inv[p_d]
                print(f"j: ({j}, {p_j}); k: ({k}, {p_k}); s: ({s}, {p_s}); d: ({d}, {p_d})")
                return p_d
            except KeyError:
                continue


```
